# Credit Card Fraud Detection dataset

О наборе данных
Контекст
Важно, чтобы компании по выпуску кредитных карт могли распознавать мошеннические транзакции, чтобы клиенты не несли расходы за товары, которые они не покупали.

Содержание
Набор данных содержит транзакции, совершенные кредитными картами в сентябре 2013 года европейскими держателями карт. Этот набор данных представляет транзакции, произошедшие за два дня, в которых зафиксировано 492 мошенничества из 284 807 транзакций. Набор данных сильно несбалансирован: положительный класс (мошенничества) составляет 0,172% всех транзакций.

Он содержит только числовые входные переменные, которые являются результатом преобразования с помощью PCA. К сожалению, из-за проблем с конфиденциальностью мы не можем предоставить исходные признаки и более подробную информацию о данных. Признаки V1, V2, … V28 — это главные компоненты, полученные с помощью PCA. Единственные признаки, которые не были преобразованы с помощью PCA, — это 'Time' и 'Amount'. Признак 'Time' содержит количество секунд, прошедших между каждой транзакцией и первой транзакцией в наборе данных. Признак 'Amount' — это сумма транзакции, этот признак может быть использован, например, для обучения с учетом затрат на примеры. Признак 'Class' — это целевая переменная, принимающая значение 1 в случае мошенничества и 0 в остальных случаях.

Учитывая соотношение классов, рекомендуется измерять точность с использованием площади под кривой точности-отзыва (AUPRC). Точность по матрице ошибок не имеет смысла для несбалансированной классификации.

```python
import torch
import torch.nn.functional as F
from torch_geometric.data import Data
from torch_geometric.nn import GCNConv
import networkx as nx
import matplotlib.pyplot as plt

# Пример транзакций
features = torch.tensor([
    [100.0],  # транзакция 0
    [200.0],  # транзакция 1
    [250.0],  # транзакция 2
    [300.0],  # транзакция 3
], dtype=torch.float)

# Метки (0 — норм, 1 — мошенник)
labels = torch.tensor([0, 1, 0, 1], dtype=torch.long)

# Рёбра графа: пары связанных транзакций (напр. по card1)
# Например: 0 связан с 1 и 2, 1 связан с 3
edge_index = torch.tensor([
    [0, 1, 0, 2, 1, 3],  # откуда
    [1, 0, 2, 0, 3, 1],  # куда
], dtype=torch.long)

# Граф PyTorch Geometric
data = Data(x=features, edge_index=edge_index, y=labels)

# ----- GNN модель -----
class GNN(torch.nn.Module):
    def __init__(self):
        super(GNN, self).__init__()
        self.conv1 = GCNConv(1, 8)
        self.conv2 = GCNConv(8, 2)  # бинарная классификация (2 класса)

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x

model = GNN()

# Обучение
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
for epoch in range(50):
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    loss = F.cross_entropy(out, data.y)
    loss.backward()
    optimizer.step()
    if epoch % 10 == 0:
        pred = out.argmax(dim=1)
        acc = (pred == data.y).float().mean()
        print(f"Epoch {epoch:02d} | Loss: {loss.item():.4f} | Acc: {acc:.2f}")

# Предсказания
model.eval()
pred = model(data.x, data.edge_index).argmax(dim=1)
print("Предсказания:", pred.tolist())
print("Истина:      ", data.y.tolist())

# Визуализация
G_nx = nx.Graph()
edges = edge_index.numpy().T
G_nx.add_edges_from(edges)

# Цвета по метке
color_map = ['green' if label == 0 else 'red' for label in data.y.tolist()]
nx.draw(G_nx, with_labels=True, node_color=color_map, node_size=600, font_color='white')
plt.title("Граф транзакций (красные = мошенники)")
plt.show()
```

# Код для подготовки графа и обучения модели (с PyTorch Geometric)

Подготовка графа и модели:

```python
import pandas as pd
import torch
from torch_geometric.data import Data
from sklearn.preprocessing import LabelEncoder, StandardScaler

# 1. Загрузка данных
df = pd.read_csv("../data/ieee-fraud-detection/train_transaction.csv")

# 2. Ограничим для демонстрации
df = df.head(10000)

# 3. Выделим сущности
categorical = ['ProductCD', 'card1', 'card2', 'card3', 'card4', 'card5', 'card6', 'addr1', 'addr2', 'P_emaildomain']
numerical = ['TransactionAmt']

# 4. Кодируем категориальные фичи
encoders = {}
for col in categorical:
    le = LabelEncoder()
    df[col] = df[col].fillna("nan")
    df[col] = le.fit_transform(df[col].astype(str))
    encoders[col] = le

# 5. Масштабируем числовые фичи
scaler = StandardScaler()
df[numerical] = scaler.fit_transform(df[numerical].fillna(0))

# 6. Формируем фичи узлов
node_features = torch.tensor(df[numerical + categorical].values, dtype=torch.float)

# 7. Метки классов
labels = torch.tensor(df['isFraud'].values, dtype=torch.long)

# 8. Граф: каждая транзакция соединена с card1, card2, ProductCD и т.д.
edge_index = []
node_id_map = {}  # храним айдишники сущностей

def get_or_add_node(entity):
    if entity not in node_id_map:
        node_id_map[entity] = len(node_id_map)
    return node_id_map[entity]

transaction_offset = 0
for i, row in df.iterrows():
    txn_id = transaction_offset + i
    for col in ['card1', 'card2', 'ProductCD', 'addr1']:
        entity_id = get_or_add_node(f"{col}_{row[col]}")
        edge_index.append([txn_id, entity_id])
        edge_index.append([entity_id, txn_id])  # bidirectional

edge_index = torch.tensor(edge_index, dtype=torch.long).T

# 9. Собираем объект графа
data = Data(x=node_features, edge_index=edge_index, y=labels)

print(data)
```

## Простая GNN модель:

```python
import torch.nn as nn
import torch.nn.functional as F
from torch_geometric.nn import SAGEConv

class GNNModel(nn.Module):
    def __init__(self, in_channels, hidden_channels):
        super().__init__()
        self.conv1 = SAGEConv(in_channels, hidden_channels)
        self.conv2 = SAGEConv(hidden_channels, 2)  # два класса

    def forward(self, x, edge_index):
        x = self.conv1(x, edge_index)
        x = F.relu(x)
        x = self.conv2(x, edge_index)
        return x
```

## Обучение:

```python
from torch_geometric.loader import DataLoader
from sklearn.model_selection import train_test_split

model = GNNModel(in_channels=data.num_node_features, hidden_channels=64)
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)
loss_fn = nn.CrossEntropyLoss()

# Разделим только по транзакционным узлам (до offset)
train_idx, val_idx = train_test_split(range(transaction_offset, transaction_offset + len(df)), test_size=0.2, random_state=42)

for epoch in range(10):
    model.train()
    optimizer.zero_grad()
    out = model(data.x, data.edge_index)
    #loss = loss_fn(out[train_idx], data.y[train_idx - transaction_offset])
    train_idx_tensor = torch.tensor(train_idx)  # if it's not already a tensor
    adjusted_idx = train_idx_tensor - transaction_offset
    loss = loss_fn(out[train_idx_tensor], data.y[adjusted_idx])

    loss.backward()
    optimizer.step()

    model.eval()
    val_pred = out[val_idx].argmax(dim=1)
    #val_true = data.y[val_idx - transaction_offset]
    val_idx_tensor = torch.tensor(val_idx)
    val_pred = out[val_idx_tensor].argmax(dim=1)
    val_true = data.y[val_idx_tensor - transaction_offset]

    acc = (val_pred == val_true).float().mean().item()
    print(f"Epoch {epoch} | Loss: {loss.item():.4f} | Val Acc: {acc:.4f}")
```

```python
import matplotlib.pyplot as plt
import networkx as nx
from torch_geometric.utils import to_networkx

# Преобразуем граф в NetworkX
G = to_networkx(data, to_undirected=True)

# Получим общее количество узлов
num_nodes = data.num_nodes

# Массив цветов
node_colors = []

for i in range(num_nodes):
    if i >= transaction_offset:
        # Это транзакция
        label_idx = i - transaction_offset
        if label_idx < len(data.y) and data.y[label_idx] == 1:
            node_colors.append('red')  # fraud
        else:
            node_colors.append('skyblue')  # обычная транзакция
    else:
        node_colors.append('lightgray')  # не транзакционный узел

# Рисуем граф
plt.figure(figsize=(12, 10))
pos = nx.spring_layout(G, seed=42)  # фиксируем расположение для повторяемости
nx.draw(G, pos, node_color=node_colors, with_labels=False, node_size=50, edge_color='lightgray')
plt.title("Граф с выделением транзакций и fraud")
plt.show()
```
